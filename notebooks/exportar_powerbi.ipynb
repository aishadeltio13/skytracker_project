```python
from pyspark.sql import SparkSession

# 1. ConfiguraciÃ³n
spark = SparkSession.builder \
    .appName("Gold_Exporter") \
    .master("local[*]") \
    .config("spark.hadoop.fs.s3a.endpoint", "http://minio:9000") \
    .config("spark.hadoop.fs.s3a.access.key", "minioadmin") \
    .config("spark.hadoop.fs.s3a.secret.key", "minioadmin") \
    .config("spark.hadoop.fs.s3a.path.style.access", "true") \
    .config("spark.hadoop.fs.s3a.impl", "org.apache.hadoop.fs.s3a.S3AFileSystem") \
    .getOrCreate()

print("ðŸš€ Iniciando exportaciÃ³n para Power BI...")

# Lista de tus KPIs
kpis = [
    "kpi_reliability_heatmap",
    "kpi_route_stress",
    "kpi_pilot_efficiency"
]

for tabla in kpis:
    # A) Leemos del sistema complejo (MinIO)
    print(f"ðŸ“¥ Leyendo {tabla}...")
    df = spark.read.parquet(f"s3a://gold/{tabla}")
    
    # B) Escribimos en sistema simple (Tu carpeta local de notebooks)
    # "file:///home/jovyan/work/..." apunta a tu carpeta "notebooks" de Windows
    ruta_local = f"file:///home/jovyan/work/datos_powerbi/{tabla}"
    
    df.write.mode("overwrite").parquet(ruta_local)
    print(f"âœ… Exportado a: notebooks/datos_powerbi/{tabla}")

print("\nðŸŽ‰ Â¡Listo! Carpeta 'notebooks/datos_powerbi'")
```
