```python
# ---------------------------------------------------------
# SCRIPT: bucket.py
# DESCRIPCIÓN: Crear bucket y guardar datos
# ---------------------------------------------------------

# 1. Instalamos la librería 'minio' DENTRO del contenedor 
!pip install minio

import os
from minio import Minio
from minio.error import S3Error

def ingest_data():
    # CONEXIÓN: Nos conectamos a "minio:9000" en vez de localhost
    client = Minio(
        "minio:9000",
        access_key="minioadmin",
        secret_key="minioadmin",
        secure=False
    )

    # CREACIÓN DE BUCKETS
    buckets = ["bronze", "silver", "gold"]
    for bucket in buckets:
        if not client.bucket_exists(bucket):
            client.make_bucket(bucket)
            print(f"Bucket creado: {bucket}")
        else:
            print(f"El bucket {bucket} ya existía")

    # SUBIDA DE ARCHIVOS
    archivos = ["flights.csv", "airlines.csv", "airports.csv"]
    
    for archivo in archivos:
        if os.path.exists(archivo):
            try:
                # Subimos a la carpeta 'raw_data' dentro de 'bronze'
                client.fput_object("bronze", f"raw_data/{archivo}", archivo)
                print(f"{archivo} subido correctamente al Data Lake")
            except S3Error as err:
                print(f"Error subiendo {archivo}: {err}")
        else:
            print(f"No encuentro {archivo}. ¿Seguro que lo moviste a la carpeta 'notebooks'?")

ingest_data()

```
